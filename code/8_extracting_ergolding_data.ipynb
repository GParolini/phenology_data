# Extracting Barley Data for the Ergolding station (Bayer)

```python
import os
from pathlib import Path
import pandas as pd
```

```python
cwd =os.getcwd()
p_cwd = Path(cwd).parent.absolute()
df = pd.read_csv(os.path.join(p_cwd, "data","phen_data.csv"), parse_dates = ['date'])
```

```python
df.head()
```

```python
df_barley_er = df[(df['s_id']==3750) & (df['species']=='Hordeum vulgare')]
```

```python
df_barley_er
```

```python
df_barley_er.to_csv(os.path.join(p_cwd, "data","barley_data", "ergolding_data.csv"), index=0)
```
