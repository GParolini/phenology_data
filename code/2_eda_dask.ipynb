```python
from pathlib import Path
import os
import pandas as pd
import dask.dataframe as dd
```

```python
cwd =os.getcwd()
p_cwd = Path(cwd).parent.absolute()
filename = "phen_data.csv"
```

```python
df = dd.read_csv(os.path.join(p_cwd, "data", filename),
                 dtype={'s_id': int,
                        'provider_id': int,
                        'lon': float,
                        'lat': float,
                        'alt': object,
                        'alt_dem': object,
                        'gss_id': int,
                        'genus': str,
                        'species': object,
                        'subspecies': object,
                        'phase_id': int,
                        'day': int,
                        'cult_season': int,
                        'affected_flag': int,
                        'qc_flag': int,
                        'qc_ori_flag': int},
                parse_dates=['date'])
```
